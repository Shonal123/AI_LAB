```python
def american(person):
    return person == "Robert"

def hostile(nation):
    return nation == "Country A"

def sells(seller, weapon, nation):
    return seller == "Robert" and weapon == "missiles" and nation == "Country A"

def is_weapon(item):
    return item == "missiles"

def is_criminal(person):

    return american(person) and is_weapon("missiles") and sells(person, "missiles", "Country A") and hostile("Country A")


name = input("Enter the name of the person to check: ")


if is_criminal(name):
    print(f"{name} is a criminal.")
else:
    print(f"{name} is not a criminal.")

```
